```python
import pandas as pd #pour gerer les bases de donn√©es et (faire la data prep)
#On va importer un premier algorithme. Logistic Regression.
from sklearn.linear_model import LogisticRegression

```

```python
train=pd.read_csv("train.csv",sep=',')
test=pd.read_csv("test.csv",sep=',')
index=test['PassengerId']
```

```python
train.head(10)
```

```python
#On indexe le dataFrame par le numero de passager. La variable n'a aucune 'valeur metier' drop = True valide cette hypoth√®se.
train.set_index('PassengerId',inplace=True,drop=True)
#On fait la m√™me chose pour le jeu de donn√©e test.csv
test.set_index('PassengerId',inplace=True,drop=True)
```

```python
test.head(10)
```

```python

class_dummies = pd.get_dummies(train['Sex'])
train=train.join(class_dummies)


class_dummies = pd.get_dummies(test['Sex'])
test=test.join(class_dummies)
```

```python
del train['Name']
del train['Ticket']
del train['Embarked']
del train['Cabin']
del train['Sex']
del train['Fare']


del test['Name']
del test['Ticket']
del test['Embarked']
del test['Cabin']
del test['Sex']
del test['Fare']

```

```python
train.head(10)
```

```python
test.head(10)
```

```python
target=train.Survived
del train['Survived']
```

```python
target.head(10)
```

```python
LR=LogisticRegression()

LR.fit(train,target)
prediction=LR.predict(test)
```

```python
Moyenne = train.Age.mean()

train['Age']= train.Age.fillna(Moyenne)
test['Age']= test.Age.fillna(Moyenne)
```

```python
LR=LogisticRegression()
LR.fit(train,target)
prediction=LR.predict(test)
```

```python
prediction
```

```python
Resultat=pd.DataFrame()
Resultat['PassengerId']=index
Resultat['Survived']=prediction
```

```python
Resultat.head(10)
Resultat.set_index('PassengerId',inplace=True,drop=True)
```

```python
Resultat.to_csv('Suvivant_Titanic.csv')
```
